### ***Task 3. Data Visualization***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df = pd.read_csv('SaYoPillow.csv')
#Renaming columns for better understanding
df.rename(columns={'sr':'snoring_range','rr':'respitory_range','t':'temperature','lm':'limb_movement_rate','bo':'oxygen_in_blood','rem':'rapid_eye_movement','sr.1':'number_of_hours_of_sleep','hr':'heart_rate','sl':'stress_level'}, inplace=True)
columns_required = ['number_of_hours_of_sleep', 'respitory_range', 'heart_rate', 'oxygen_in_blood', 'rapid_eye_movement']

# To avoid truncation of OP
pd.set_option('display.max_colwidth', None)
# Ensuring the DataFrame only contains the columns of interest
# df = df[columns_required]
output_column = 'stress_level'
df = df[columns_required + [output_column]]
df.head()
```

***Task 3.1: Create a boxplot (or box-and-whisker plot) for each variable (X1;X2;X3;X4;X5 ). These plots will highlight the median, quartiles, and outliers, providing a concise overview of each variable's distribution.***

```python
# plt.figure(figsize=(12, 8))
for i, var in enumerate(columns_required):
    sns.boxplot(x=df[var])
    plt.title(f'Boxplot of {var}')
    # plt.tight_layout()
    plt.show()
```

***Task 3.2: Plot Scatter plots to display the relationship between the output data column Y with respect to each variable (i.e., the column)`.***

```python
# plt.figure(figsize=(12, 10))

for var in columns_required:
    sns.scatterplot(x=df[var], y=df[output_column])
    plt.title(f'{var} vs. {output_column}')
    plt.xlabel(var)
    plt.ylabel(output_column)
    plt.show()
```

***Task 3.3: Plot the density curve of the output data.***

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df['stress_level'], fill=True)
plt.title('Density Curve of Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Density')
plt.show()
```

***(c) Reflect on how these visualizations enhance the understanding of the dataset and the analysis process.***

**Facilitating Communication**: They are a powerful tool for communicating findings to others, particularly those who may not be familiar with statistical terminology or methods.

**Guiding Further Analysis**: Visualizations can guide further analysis by highlighting areas that need more in-depth investigation or by suggesting appropriate statistical models.

**Enhancing Interpretation**: By making patterns in the data visible, visualizations enhance the interpretation of the data and can lead to insights that might be missed if only raw data or summary statistics were considered.

**Efficiency**: Visual analysis can often be more efficient than purely numerical analysis, allowing for quicker detection of trends and outliers.

**Holistic View**: Visualizations provide a holistic view of the data, capturing multiple aspects of the data distribution at once.

**Understanding Distribution**: Visualizations provide a smooth estimate of the distribution, showing where values are concentrated over an interval.

**Identification of Skewness**: They help in identifying whether the data is symmetric or skewed to one side, which is vital for understanding the nature of the data and for choosing appropriate statistical tests.

**Relationships Between Variables**: Visualizations are essential to investigate the potential relationships between variables. They can reveal correlation patterns, whether they are linear or non-linear, and the strength of the relationship.

**Outliers and Anomalies**: Scatter plots can also reveal outliers or anomalies that deviate significantly from the general pattern, which could be influential points or data entry errors.
