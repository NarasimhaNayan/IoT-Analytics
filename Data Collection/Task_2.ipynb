### ***Code Calculation***

```python
# Finding out mean, median, varience, range, IQR using the functions available in pandas
print('Task 2.1: Descriptive Statistics for the 5 variables considered: ')

statistics = pd.DataFrame(index=['mean', 'median', 'variance', 'range', 'IQR'])

for column in columns_required:
    statistics.loc['mean', column] = df[column].mean()
    statistics.loc['median', column] = df[column].median()
    statistics.loc['variance', column] = df[column].var()
    statistics.loc['range', column] = df[column].max() - df[column].min()
    statistics.loc['IQR', column] = df[column].quantile(0.75) - df[column].quantile(0.25)

print(statistics)
```

***Task 2.2 Probability Mass Function (PMF): For each variable Xi*** 

***(a) segment the range of the dataset into 50 equal intervals, ensuring that the minimum and maximum values of the dataset are included.*** 

***(b) Plot the PMF for each variable Xi . The PMF will exhibit the probability distribution of the discrete variables in the dataset***

```python
# pmfs = {}

for var in columns_required:
    # Determine the bin edges for 50 equal intervals, including the exact range of the data
    min_val = df[var].min()
    max_val = df[var].max()
    bins = np.linspace(min_val, max_val, num=51)
    
    # a. Segmenting
    # Use pd.cut to bin the data
    binned_data = pd.cut(df[var], bins=bins, include_lowest=True, right=True)
    print('Segmentation of', var,':')
    # print(binned_data)

    # Calculate the frequency of observations in each bin
    freq = binned_data.value_counts(sort=False, normalize=True)
    #  b. Plotting PMF
    plt.figure(figsize=(12, 6))
    freq.plot(kind='bar')
    plt.title(f'PMF of {var}')
    plt.xlabel('Interval')
    plt.ylabel('Probability')
    plt.xticks(rotation=45)
    plt.show()


    print("\n")
```

2. Outliers: Identify and discuss any outliers present in the dataset. Evaluate their potential impact on the overall analysis.


```python
outliers = {}

for var in columns_required:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    
    # Identifying outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter the DataFrame rows where var is outside the lower and upper bounds
    outliers[var] = df[(df[var] < lower_bound) | (df[var] > upper_bound)]
    
    # You can also directly calculate the number of outliers
    num_outliers = ((df[var] < lower_bound) | (df[var] > upper_bound)).sum()
    print(f"{var} has {num_outliers} outliers")
    print('Outliers for :',var)
    print(outliers[var])
    
# outliers dictionary now contains the outlier rows for each variable
```

Analysis of Outliers:

As we observe that all the dataframes are empty. This indicates that the dataset has no outliers for the columns that are considered. Having no outliers implicates that the data is clean and well curated.

4. Spread: Analyze the variability of the data, using measures such as range, IQR, and variance,for each variable Xi . The report should elucidate a comprehensive understanding of each variable's statistical properties and their implications on the output variable Y .

```python
for var in columns_required:    
    range_val = df[var].max() - df[var].min()
    iqr_val = np.subtract(*np.percentile(df[var], [75, 25]))
    variance_val = df[var].var()
    print(f'{var} - Range: {range_val}, IQR: {iqr_val}, Variance: {variance_val}\n')
```
